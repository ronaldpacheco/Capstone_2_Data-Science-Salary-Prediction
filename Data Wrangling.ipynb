```python
#Let's write a function to compare all the column names of two given dataframes
def check_columns(df1, df2):
    columns1 = df1.columns
    columns2 = df2.columns
    matches = df1.columns == df2.columns
    if np.sum(matches) == len(matches):
        print("All columns are the same")
    else:
        print(columns1[~matches])
```

Here we can see all the variables that are missing from each dataset. We already stablished that some of them need to be broken up, and after that, we need to decide what to do with the variables that were not present on both sets of dataframes from the beginning.

Let's start with `Location`:

```python
def location(df):
    df.Location = df.Location.str.split(',')
    df['City'] = [i[0] for i in df.Location]
    df['State'] = [i[1].strip() for i in df.Location]
    df.drop(columns = 'Location', inplace=True)
    return df
```

Now let's break up `Salary Estimate`:

```python
#Function to split 'Salary Estimate into 'Min_Salary' and 'Max_Salary'. Also removes '$' and 'K', turns the variable type into integer and multiplies by 1000
def salary(df):
    df['Salary Estimate'] = df['Salary Estimate'].str.split('-')
    df['Min_Salary'] = [i[0].strip('$') for i in df['Salary Estimate']]
    df['Max_Salary'] = [i[1].strip('$') for i in df['Salary Estimate']]
    df['Max_Salary'] = df['Max_Salary'].str.split('(')
    df['Glassdoor_Estimate'] = [0 if i[1].find('Glass') == -1 else 1 for i in df['Max_Salary']]
    df['Max_Salary'] = [i[0].strip() for i in df['Max_Salary']]
    df.drop(columns = 'Salary Estimate', inplace=True)
    df['Min_Salary'] = df.Min_Salary.str.strip('K').astype(int)*1000
    df['Max_Salary'] = df.Max_Salary.str.strip('K').astype(int)*1000
    return df
```
