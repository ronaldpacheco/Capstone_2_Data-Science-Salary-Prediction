Let's start with `Location`:

```python
def location(df):
    df['In_Headquarters'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
    df.Location = df.Location.str.split(',')
    df['City'] = [i[0] for i in df.Location]
    df['State'] = [i[1].strip() for i in df.Location]
    #Reverting the split so we don't encounter issues with df operations
    df['Location'] = df['Location'].str.join(',')
    #df.drop(columns = 'Location', inplace=True)
    return df
```

Now let's break up `Salary Estimate`:

```python
#Function to split 'Salary Estimate into 'Min_Salary' and 'Max_Salary'. Also removes '$' and 'K', turns the variable type into integer and multiplies by 1000
def salary(df):
    df['Salary Estimate'] = df['Salary Estimate'].str.split('-')
    df['Min_Salary'] = [i[0].strip('$') for i in df['Salary Estimate']]
    df['Max_Salary'] = [i[1].strip('$') for i in df['Salary Estimate']]
    df['Max_Salary'] = df['Max_Salary'].str.split('(')
    df['Employer_Estimate'] = [0 if i[1].find('mploy') == -1 else 1 for i in df['Max_Salary']]
    df['Max_Salary'] = [i[0].strip() for i in df['Max_Salary']]
    #Reverting the split so we don't encounter issues with df operations
    df['Salary Estimate'] = df['Salary Estimate'].str.join('-')
    #df.drop(columns = 'Salary Estimate', inplace=True)
    df['Min_Salary'] = df.Min_Salary.str.strip('K').astype(int)*1000
    df['Max_Salary'] = df.Max_Salary.str.strip('K').astype(int)*1000
    return df
```

First, let's create or final response variable `Avg_Salary`:

Now, we want our imputed observations to be good estimation of the real variable, that is: if our `Size` is 10000+, our salary might tend to be higher as bigger companies tend to pay higher salaries. Also, a Data Analyst will usually make less than a Senior Data Scientist, but maybe a Data Analyst working for a company with 100,000+ employees and a `Revenue` of 10+ billion (USD) will earn a similar amount to a Senior Data Scientist in a company with 50 employees and less than 1 million (USD) in revenue... So in order to impute these variables, we need to define our categories as dummy variables using `Job_title` and `Job_Desc`.

We want to categorize `Job_title` by: Position: Analyst, Engineer, Scientist, Deep (Learning), Machine (Learning). Senority: Manager, Senior or Sr., Junior or Jr., Lead, Principal, Associate, Chief, Intern.

```python
df['Title_Analyst'] = df.apply(lambda x: 1 if 'analyst' in x.Job_title.lower() else 0, axis=1)
df['Title_Engineer'] = df.apply(lambda x: 1 if 'engineer' in x.Job_title.lower() else 0, axis=1)
df['Title_Scientist'] = df.apply(lambda x: 1 if 'scientist' in x.Job_title.lower() else 0, axis=1)
df['Title_Deep'] = df.apply(lambda x: 1 if 'deep' in x.Job_title.lower() else 0, axis=1)
df['Title_Machine'] = df.apply(lambda x: 1 if 'machine' in x.Job_title.lower() else 0, axis=1)
df['Title_Manager'] = df.apply(lambda x: 1 if 'manager' in x.Job_title.lower() else 0, axis=1)
df['Title_Senior'] = df.apply(lambda x: 1 if 'senior' in x.Job_title.lower() or 'sr' in x.Job_title.lower() else 0, axis=1)
df['Title_Junior'] = df.apply(lambda x: 1 if 'junior' in x.Job_title.lower() or 'jr' in x.Job_title.lower() else 0, axis=1)
df['Title_Lead'] = df.apply(lambda x: 1 if 'lead' in x.Job_title.lower() else 0, axis=1)
df['Title_Principal'] = df.apply(lambda x: 1 if 'principal' in x.Job_title.lower() else 0, axis=1)
df['Title_Chief'] = df.apply(lambda x: 1 if 'chief' in x.Job_title.lower() else 0, axis=1)
df['Title_Intern'] = df.apply(lambda x: 1 if 'intern' in x.Job_title.lower() else 0, axis=1)
```
